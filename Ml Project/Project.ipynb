```python
def utils_recognize_type(data, col, max_cat=20):
    if (data[col].dtype == "O") | (data[col].nunique() < max_cat):
        return "cat"
    else:
        return "num"

#heatmap of variables
dic_cols = {col:utils_recognize_type(data, col, max_cat=20) for col in data.columns}
print(dic_cols)
```

```python
for x in data.columns:
    print(x)
    if(dic_cols[x]=='num'):
        print("Mean ="+ str(data[x].mean()))
        print("Variance ="+ str(data[x].var()))
        print("Standard Deviation ="+ str(data[x].std()))
    else:
        print("Median ="+ str(data[x].median()))
        print("Mode ="+ str(data[x].mode().iat[0]))
    print('\n')
```

```python
X = data.drop(['PowerAll'],axis=1).values
Y = data['PowerAll'].values
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.3,random_state=2)
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train,y_train)
```
