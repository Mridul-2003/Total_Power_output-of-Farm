```python
def utils_recognize_type(data, col, max_cat=20):
    if (data[col].dtype == "O") | (data[col].nunique() < max_cat):
        return "cat"
    else:
        return "num"

#heatmap of variables
dic_cols = {col:utils_recognize_type(data, col, max_cat=20) for col in data.columns}
print(dic_cols)
```

```python
print(data.corrwith(data['PowerAll']).sort_values(ascending=False))
```

```python
X = data.drop(['PowerAll'],axis=1).values
Y = data['PowerAll'].values
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.3,random_state=2)
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression,Lasso
model = LinearRegression()
model2 = Lasso()
model.fit(x_train,y_train)
model2.fit(x_train,y_train)
```

```python
y_predict_train = model.predict(x_train)
```

```python
import seaborn as sns
ax1 = sns.distplot(y_train, hist=False, color="r", label="Actual Value")
sns.distplot(y_predict_train, hist=False, color="b", label="Fitted Values" , ax=ax1)
```

```python
import matplotlib.pyplot as plt
plt.scatter(y_train,y_predict_train)
plt.plot(y_train,y_predict_train,c='r')
plt.xlabel("Actual")
plt.ylabel("Predict")
```
